```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('fffff.csv')
df1
```

```python
df1.info()
```

```python
##a) Los registros de 5 “host name” diferentes
cinco_hosts = df1["host_name"].dropna().unique()[:5]
print(f"nombre de los cinco hosts diferentes:{cinco_hosts}")
```

```python
##b) Los registros de los host que se hallan unido a Airbnb “despues del año 2020”
df1["host_since"] = pd.to_datetime(df1["host_since"], format="%d/%m/%Y")
filtrob = df1[df1["host_since"] > "2020-12-31"]
filtrob[["host_name", "host_since"]].head(5)
```

```python
##c) Los registros de los host que responden “a más tardar en 1 día”
filtroc = df1[df1["host_response_time"] == "within a day"]
filtroc[["host_name", "host_response_time"]].head(5)
```

```python
##d) Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)
filtrod = df1.iloc[::200]
filtrod.head()
```

```python
##e) Los registros de los tipos de cuarto: “Entire home” y “Private room
filtroe = df1[df1["room_type"].isin(["Entire home/apt", "Private room"])]
filtroe.head()
```

```python
##f) Los registros que “a lo mucho cuenten con 4 camas
filtrof = df1[df1["beds"] <= 4]
filtrof.head()
```

```python
##g) Los registros que “sean superhost”
filtrog = df1[df1["host_is_superhost"] == "t"]
filtrog.head()
```

```python
##h) Los registros de las columnas “impares”
filtroh = df1.iloc[:, ::2]  
filtroh.head()
```
