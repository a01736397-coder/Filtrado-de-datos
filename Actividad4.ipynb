```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Datos_Facturación.csv')
df
```

```python
df.info()
```

```python
## ejercicio 1  CVE_CLPV entre 1000 y 2000
df1 = df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
df1
```

```python
df1.to_csv("df1.csv")
```

```python
## Ejercicio 2 todas las claves excepto 5 y 4
df2 = df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]
df2
```

```python
df2.to_csv("df2.csv")
```

```python
##ejercicio 3 solo 28/02/2022 
df3 =df[(df["FECHA_ENT"] == "2022-02-28 00:00:00")]
df3
```

```python
df3.to_csv("df3.csv")
```

```python
##ejercicio 4 CAN_TOT < 5951.7  OR  STATUS == "E"
df4 = df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
df4
```

```python
df4.to_csv("df4.csv")
```

```python
##ejercicio 5 solo columnas: CVE_DOC, FECHA_ENT, FECHA_VEN, CAN_TOT
df5 = df[["CVE_DOC", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]
df5
```

```python
df5.to_csv("df5.csv")
```

```python
##Ejercicio 6 solo filas 7001–7099 
df6 = df.iloc[7000:7100]
df6
```

```python
df6.to_csv("df6.csv")
```

```python
##Ejercicio 7 index = CVE_VEND(valores 1,2) (columna: FECHALAB)
df7 = df[df["CVE_VEND"].isin([1, 2])].set_index("FECHAELAB")
df7
```

```python
df7.to_csv("df7.csv")
```
